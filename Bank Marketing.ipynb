Checking for Outliers

```python
#Check the outliers in numerical columns
plt.figure(figsize=(14,10),facecolor="white")
plotnumber=1
for col in numerical_col:
    ax=plt.subplot(3,3,plotnumber)
    sns.boxplot(df[col],palette="bright")
    plt.xlabel(col,fontsize=15)
    plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

We can find a lot of outliers in all the columns except the column "day". So, we will remove these outliers using Z-Score technique.

```python
#Remove outliers
from scipy.stats import zscore
out_features=df.drop(categorical_col+["day"],axis=1)
z=np.abs(zscore(out_features))
df1=df[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
print("Data loss percentage - ",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

The data loss percentage is 11.13% which is more than 10% and hence not acceptable. So, we will continue with the original dataset "df" without removing the outliers.

Encoding Categorical Columns

```python
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder()
for col in categorical_col:
    df[col]=OE.fit_transform(df[col].values.reshape(-1,1))
df
```

Checking for Skewness

```python
df.skew()
```

We can see that there is a lot of skewness in the features "age", "default", "balance", "loan", "contact", "duration", "campaign", "pdays", "previous" and "poutcome". But the features "default", "loan", "contact" and "poutcome" are categorical in nature so no need to remove skewness in them. We will try to reduce skewness from the continuous features using Yeo-Johnson transformation method.

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
df["age"]=pt.fit_transform(df[["age"]])
df["balance"]=pt.fit_transform(df[["balance"]])
df["duration"]=pt.fit_transform(df[["duration"]])
df["campaign"]=pt.fit_transform(df[["campaign"]])
df["pdays"]=pt.fit_transform(df[["pdays"]])
df["previous"]=pt.fit_transform(df[["previous"]])
```

```python
#Check for skewness again
df.skew()
```

Now, we can see that the skewness has been reduced a lot in the features.

Separating features and label

```python
x=df.drop('subscribed',axis=1)
y=df['subscribed']
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

Now, we have done feature scaling to overcome the issue of biasness in the model.

Checking Variance Inflation Factor (VIF)

```python
#Find Variance Inflation Factor of each feature
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

We check VIF values to find out if there is any multicollinearity problem with the features which we need to solve. From the above table, we can see that the features "pdays" and "previous" have VIF values greater than 10. We will first remove the feature "pdays" which has the highest VIF value and then check the VIF values again.

```python
x.drop("pdays",axis=1,inplace=True)
```

```python
#Check VIF values again
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif["Features"]=x.columns
vif
```

Now, we can see that none of the features have VIF values greater than 10. So, we have solved the problem of multicollinearity and can proceed further.

```python
y.value_counts()
```

We can see from the above values that the data in the label is not balanced. As this is a classification problem, we need to balance the data.

Balancing the Label

```python
#Oversampling the data using SMOTE
from imblearn.over_sampling import SMOTE
SM=SMOTE()
x1,y1=SM.fit_resample(x,y)
```

```python
y1.value_counts()
```

```python
x1.shape
```

# Model Building

Finding the best random state

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    RFC=RandomForestClassifier()
    RFC.fit(x_train,y_train)
    pred=RFC.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu,"at random state",maxRS)
```

Creating train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=maxRS)
```

Importing the necessary modules

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import ExtraTreesClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#Random Forest Classfier
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
#Logistic Regression
LR=LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
print(classification_report(y_test,predLR))
```

```python
#XGB Classifier
XGB=XGBClassifier()
XGB.fit(x_train,y_train)
predXGB=XGB.predict(x_test)
print(accuracy_score(y_test,predXGB))
print(confusion_matrix(y_test,predXGB))
print(classification_report(y_test,predXGB))
```

```python
#KNeighbors Classifier
KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
predKNN=KNN.predict(x_test)
print(accuracy_score(y_test,predKNN))
print(confusion_matrix(y_test,predKNN))
print(classification_report(y_test,predKNN))
```

```python
#Extra Trees Classifier
ET=ExtraTreesClassifier()
ET.fit(x_train,y_train)
predET=ET.predict(x_test)
print(accuracy_score(y_test,predET))
print(confusion_matrix(y_test,predET))
print(classification_report(y_test,predET))
```

```python
#Gradient Boosting Classifier
GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
predGB=GB.predict(x_test)
print(accuracy_score(y_test,predGB))
print(confusion_matrix(y_test,predGB))
print(classification_report(y_test,predGB))
```

```python
#AdaBoost Classifier
ABC=AdaBoostClassifier()
ABC.fit(x_train,y_train)
predABC=ABC.predict(x_test)
print(accuracy_score(y_test,predABC))
print(confusion_matrix(y_test,predABC))
print(classification_report(y_test,predABC))
```

```python
#Bagging Classifier
BC=BaggingClassifier()
BC.fit(x_train,y_train)
predBC=BC.predict(x_test)
print(accuracy_score(y_test,predBC))
print(confusion_matrix(y_test,predBC))
print(classification_report(y_test,predBC))
```

# Hyper Parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param={'n_neighbors': np.arange(1, 31, 1),
       'weights': ['uniform', 'distance'],
       'metric': ['euclidean', 'manhattan', 'minkowski'],
       'p': np.arange(1, 6)  #Common values for p in Minkowski metric are 1, 2, 3, 4, 5
      }
RCV=RandomizedSearchCV(KNeighborsClassifier(),param,cv=5)
RCV.fit(x_train,y_train)
```

```python
#Best parameters for the final model
RCV.best_params_
```

# Building the Final Model

```python
final_model=KNeighborsClassifier(n_neighbors=2,weights='uniform',metric='minkowski',p=2)
final_model.fit(x_train,y_train)
pred=final_model.predict(x_test)
acc=accuracy_score(y_test,pred)
print("Accuracy acheived by the final model is ",acc*100)
```

Plotting ROC and comparing AUC for the final model

```python
from sklearn import metrics
fpr,tpr,thresholds=metrics.roc_curve(y_test,pred)
roc_auc=metrics.auc(fpr,tpr)
display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=final_model)
display.plot()
plt.show()
```

Above is the ROC curve for the final model and AUC value for the final model is 94%.

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
df1["age"]=pt.fit_transform(df1[["age"]])
df1["balance"]=pt.fit_transform(df1[["balance"]])
df1["duration"]=pt.fit_transform(df1[["duration"]])
df1["campaign"]=pt.fit_transform(df1[["campaign"]])
df1["pdays"]=pt.fit_transform(df1[["pdays"]])
df1["previous"]=pt.fit_transform(df1[["previous"]])
```

Now, we can see that the skewness has been reduced a lot in the features.

Checking Variance Inflation Factor (VIF)

```python
#Find Variance Inflation Factor of each feature
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(df1.values,i) for i in range(len(df1.columns))]
vif["Features"]=df1.columns
vif
```

We check VIF values to find out if there is any multicollinearity problem with the features which we need to solve. From the above table, we can see that the features "pdays" and "previous" have VIF values greater than 10. We will first remove the feature "pdays" which has the highest VIF value and then check the VIF values again.

```python
df1.drop("pdays",axis=1,inplace=True)
```

```python
#Check VIF values again
vif=pd.DataFrame()
vif["VIF Values"]=[variance_inflation_factor(df1.values,i) for i in range(len(df1.columns))]
vif["Features"]=df1.columns
vif
```

Now, we can see that none of the features have VIF values greater than 10. So, we have solved the problem of multicollinearity and can proceed further.
